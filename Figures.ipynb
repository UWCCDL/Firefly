```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme("whitegrid")
```

Efficient approximation of $A(m, t)$:

$$A(m,t) =\log\left( \frac{n}{1 - d}\right) - d \times \log(t)$$ 

```python
def efficient_activation(time, n=10, d=0.5):
    return np.log(n / (1 - d)) - d*np.log(time)

x = np.linspace(1, 60*60*24*360, 500)
y10 = efficient_activation(x, n=1)
y20 = efficient_activation(x, n=10)
y30 = efficient_activation(x, n=100)


fig, ax = plt.subplots(1,1) 

ax.plot(x, y10)
ax.plot(x, y20)
ax.plot(x, y30)

ax.legend(labels=["$n = %d$" % (x) for x in [1 , 10, 100]])
ax.set_xticks([60*60*24*30 * x for x in range(0,13)])
ax.set_xticklabels(range(0,13))
ax.set_xlabel("Months Since Encoding")
ax.set_ylabel("Memory Activation")
plt.show()
```
